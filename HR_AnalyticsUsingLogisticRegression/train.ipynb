```python
data.drop(['education'],axis=1,inplace=True)
department=pd.get_dummies(data['department'],drop_first=True)
recruitment_channel=pd.get_dummies(data['recruitment_channel'],drop_first=True)
gender=pd.get_dummies(data['gender'],drop_first=True)
region=pd.get_dummies(data['region'],drop_first=True)
```

```python
data.drop(['department','recruitment_channel','gender','employee_id','region'],axis=1,inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
def rating(cols):
    previous_year_rating=cols[0]
    length_of_service=cols[1]
    if pd.isnull(previous_year_rating):
        if length_of_service==1:
            return 0.0
    else:
            return previous_year_rating
```

```python
data['previous_year_rating']=data[['previous_year_rating','length_of_service']].apply(rating,axis=1)
```

```python
data.isnull().sum()
```

```python
data['previous_year_rating'].value_counts()
```

```python
data.isnull().sum()
```

```python
data["previous_year_rating"]= data["previous_year_rating"].astype(int)
```

```python
data_dummy=pd.concat([department,recruitment_channel,region,gender,data],axis=1)
```

```python
data_dummy
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X=data_dummy.iloc[:,0:51]
y=data_dummy.iloc[:,51:52]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(X_train,np.array(y_train).ravel())
```

```python
logreg.score(X_train,y_train)
```

```python
logreg.score(X_test,y_test)
```
